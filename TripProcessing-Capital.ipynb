```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
```

```python
path = 'tripdata/capital-bikeshare/'
tripdataFilenames = [f for f in listdir(path) if isfile(join(path, f))]
print(tripdataFilenames)
print(len(tripdataFilenames))
```

```python
dtype={'trip_id': int, 'starttime': str, 'stoptime': str, 'start_time': str, 'end_time': str, 'bikeid': int, 'tripduration': int, 
       'from_station_id': int, 'from_station_name': str, 'to_station_id': int, 'to_station_name': str, 
       'usertype': str, 'gender': str, 'birthday': str, 'birthyear': str}

tripdataFrames = [
    pd.read_csv(path + filename, dtype=dtype).rename(columns = {'start_time':'starttime', 'end_time':'stoptime', 'birthday':'birthyear'}) 
    for filename in tripdataFilenames]
```

```python
tripdataFrame = pd.concat(tripdataFrames)
tripdataFrame = tripdataFrame.rename(columns = {'from_station_id':'start station id', 'to_station_id':'end station id'})
```

```python
tripdataFrame
```

```python
frequencyByStation = tripdataFrame[[' start station id',' end station id',' starttime']].groupby([' start station id',' end station id']).count()
```

```python
frequencyByStation.to_csv('capital-bikeshare_frequencyByStation.csv')
```
