```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
```

```python
path = 'tripdata/hubway/'
tripdataFilenames = [f for f in listdir(path) if isfile(join(path, f))]
```

```python
tripdataFrames = [pd.read_csv('tripdata/hubway/' + filename) for filename in tripdataFilenames]
```

```python
tripdataFrame = pd.concat(tripdataFrames)
```

```python
tripdataFrame
```

```python
frequencyByStation = tripdataFrame[['start station id','end station id','starttime']].groupby(['start station id','end station id']).count()
```

```python
frequencyByStation.to_csv('hubway_frequencyByStation.csv')
```
